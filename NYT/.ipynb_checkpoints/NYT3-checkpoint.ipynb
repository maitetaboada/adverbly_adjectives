```python
import requests
import pandas as pd
import numpy as np
import sys, os, re, time
import json, csv
from bs4 import BeautifulSoup
from collections import defaultdict
```

```python
df = pd.read_csv('nyt_movie_reviews.csv', parse_dates=['opening_date', 'publication_date', 'date_updated'])
```

```python
os.makedirs('nyt_articles/', exist_ok=True)

for index,row in df.iterrows():
    if index in [515, 4, 5, 6, 7, 8, 10, 267, 15, 19, 278, 279, 26, 289, 91, 262, 689, 297, 300, 301, 302, 47, 520, 563, 309, 55, 570, 578, 70, 73, 331, 45, 590, 591, 80, 338, 84, 85, 86, 88, 527, 93, 606, 608, 98, 616, 619, 365, 370, 117, 632, 633, 635, 381, 639, 128, 'index', 387, 133, 646, 391, 650, 395, 652, 142, 401, 664, 661, 152, 282, 158, 160, 680, 169, 172, 685, 686, 687, 688, 177, 178, 179, 182, 183, 696, 705, 587, 454, 545, 205, 719, 724, 475, 223, 224, 482, 234, 493, 238, 497, 242, 244, 246, 249, 252, 605]:
    
        article_url = row['link']
        print(article_url)
        r = requests.get(article_url)

        soup = BeautifulSoup(r.text, 'html.parser')

        articlebody = soup.find('section', {'name' : 'articleBody'})

        if articlebody:
            with open('nyt_articles/' + str(index) + '.txt', 'w+', encoding='utf-8') as fo:
                for p in articlebody.findAll('p'):
                    fo.write(p.text + '\n')

        else:
            articlebody = soup.find('article', {'id' : 'story'})
            if articlebody:
                with open('nyt_articles/' + str(index) + '.txt', 'w+', encoding='utf-8') as fo:
                    for p in articlebody.findAll('p', {'class' : 'story-body-text', 'class' : 'story-content'}):
                        fo.write(p.text + '\n')
```
