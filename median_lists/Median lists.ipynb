```python
import numpy as np
import pandas as pd
import os, sys
import matplotlib.pyplot as plt
```

```python
def makeDirs(dirs):
    curr_path = '.'
    for d in dirs:
        os.makedirs(curr_path+'/'+d, exist_ok=True)
        curr_path = curr_path+'/'+d
```

```python
def getMedianLists(filepath):
    split = filepath.split('/')
    fname = split[-1]
    dir_structure = ['median_lists']
    dir_structure += split[1:-1]
    makeDirs(dir_structure)
    median_path = '/'.join(dir_structure) + '/' + fname
    print(median_path)
    
    df = pd.read_csv(filepath)
    df = df[df['counts'] != 1]
    num_pairs = len(df)
    
    # true median
    middle = df['counts'].median()
    
    df.loc[middle-100:middle+99].to_csv(median_path, index=None)
```

```python
def getMiddleLists(filepath):
    split = filepath.split('/')
    fname = split[-1]
    dir_structure = ['middle_lists']
    dir_structure += split[1:-1]
    makeDirs(dir_structure)
    middle_path = '/'.join(dir_structure) + '/' + fname
    print(middle_path)
    
    df = pd.read_csv(filepath)
    df = df[df['counts'] != 1]
    num_pairs = len(df)
    
    # middle of distribution
    middle = round(num_pairs / 2)
    
    df.loc[middle-100:middle+99].to_csv(middle_path, index=None)
```

```python
basepath = 'adverbly_adjectives'
files = []
for corpus in os.listdir(basepath):
    for f_or_d in os.listdir(basepath+'/'+corpus):
        if os.path.isdir(basepath+'/'+corpus+'/'+f_or_d):
            genres_list = [basepath+'/'+corpus+'/'+f_or_d+'/'+f for f in os.listdir(basepath+'/'+corpus+'/'+f_or_d)]
            files += genres_list
        else:
            assert os.path.isfile(basepath+'/'+corpus+'/'+f_or_d)
            file = [basepath+'/'+corpus+'/'+f_or_d]
            files += file
            
for f in files:
#     print(f)
    getMedianLists(f)
    getMiddleLists(f)
```
