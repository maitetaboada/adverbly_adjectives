```python
import os, re
import csv
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
freq_df = pd.DataFrame(pd.read_excel('COCA_top500_annotated_sorted_frequency.xlsx', sheet_name=1, header=None, usecols=[2,4,5], names=['pair', 'type', 'additional']))
freq_df['pair'] = freq_df['pair'].str.lower().str.strip()

genre_freq_df = pd.DataFrame(pd.read_excel('COCA_top500_annotated_sorted_frequency.xlsx', sheet_name=1, header=None, usecols=[2,4,5], names=['pair', 'type', 'additional']))
genre_freq_df['pair'] = genre_freq_df['pair'].str.lower().str.strip()

print(len(freq_df))
freq_df.head()
```

```python
basepath = 'adverbly_adjectives'
files = []
for corpus in os.listdir(basepath):
    for f_or_d in os.listdir(basepath+'/'+corpus):
        
        # handle genres
        if os.path.isdir(basepath+'/'+corpus+'/'+f_or_d):
            genres_list = [basepath+'/'+corpus+'/'+f_or_d+'/'+f for f in os.listdir(basepath+'/'+corpus+'/'+f_or_d)]

            for f in genres_list:
                genre = f.split(f_or_d+'/')[1].split('_pairs.csv')[0]
                print(corpus, genre)
                suffix = '_' + corpus + '_' + genre
                df = pd.read_csv(f)
                genre_freq_df = genre_freq_df.merge(df, on='pair', how='left').fillna(0).rename({'counts' : 'counts' + suffix, 'freqmil' : 'freqmil'+suffix}, axis=1)
                
        # handle master lists for each corpus
        else:
            suffix = '_' + corpus
            print(corpus)
            assert os.path.isfile(basepath+'/'+corpus+'/'+f_or_d)
            file = basepath+'/'+corpus+'/'+f_or_d
            df = pd.read_csv(file)
            freq_df = freq_df.merge(df, on='pair', how='left').fillna(0).rename({'counts' : 'counts' + suffix, 'freqmil' : 'freqmil'+suffix}, axis=1)
```
